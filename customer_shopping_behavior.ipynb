```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv("customer_shopping_behavior.csv")
```

```python
df['Review Rating']= df.groupby('category')['review_rating'].transform(lambda x:x.fillna(x.median()))
```

```python
df.isnull().sum()
```

```python
df.columns = df.columns.str.lower()
df.columns = df.columns.str.replace(' ','_')
df = df.rename(columns={'purchase_amount_(usd)':'purchase_amount'})
```

```python
df.info()
```

```python
labels= ['Young_adult','Adult','Middle_age','Senior']
df['age_group']= pd.qcut(df['age'],q=4, labels=labels)
```

```python
df[['age','age_group']].head(10)
```

```python
# create new column purchase_frequency_days

frequency_mapping = {
    'Fortnightly': 14,
    'Weekly': 7,
    'Monthly': 30,
    'Quarterly': 90,
    'Bi-Weekly': 14,
    'Annually': 365,
    'Every 3 Months': 90
}
df['purchase_frequency_day']=df['frequency_of_purchases'].map(frequency_mapping)
```
